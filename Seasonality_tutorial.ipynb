```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

## based on https://towardsdatascience.com/how-to-find-seasonality-using-python-73547ebf322

## We will analyize the number of crimes (breaking and entering) per month.


## Dataset from:
## https://data.kcmo.org/dataset/Kansas-City-Crime-NIBRS-Summary/6wc4-sd7p

# Mount Google Drive in Google Colab (if applicable)
from google.colab import drive
drive.mount('/content/drive')

# Original path: 'https://drive.google.com/file/d/1KXiFUXH9JJ4MXi84o34zZi7MPc3N01ov/view?usp=sharing'
# Updated path using the raw file URL
path = 'https://drive.google.com/uc?id=1KXiFUXH9JJ4MXi84o34zZi7MPc3N01ov&export=download'

df = pd.read_csv(path)
df.head()
```

```python
df = df[['Date','Burglary/Breaking and Entering']]
df.head()
```

```python
## We normalize the data by dividing the number of crimes per month by the number of days

df['Burglary/Breaking and Entering'] = df['Burglary/Breaking and Entering']/(pd.to_datetime(df['Date']).dt.day)
```

```python
## We set the index to be the date column.

## We will create a helper function

def set_date_index(input_df, col_name = 'Date'):

  modified_df = input_df.copy()

  modified_df[col_name] = pd.to_datetime(modified_df[col_name])

  modified_df.sort_values(col_name, inplace = True)
  modified_df.set_index(col_name, inplace = True)

  return modified_df
```

```python
df = set_date_index(df,'Date')
df
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Replace 'freq' with 'period'
sd = seasonal_decompose(df, period=12)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Replace 'freq' with 'period'
sd = seasonal_decompose(df, period=12)


def combine_seasonal_cols(input_df, sd_object):
    """Combines seasonal components with the original DataFrame.
    """
    modified_df = input_df.copy()
    modified_df['seasonal'] = sd_object.seasonal
    modified_df['trend'] = sd_object.trend
    modified_df['resid'] = sd_object.resid

    return modified_df

combine_seasonal_cols(df,sd)
```

```python
df = combine_seasonal_cols(df, sd)
plt.figure(figsize=(14,7))
plt.plot(df['seasonal'],color='red')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.plot(df['trend'],color='red')
# Assuming the original time series data is in a column named 'Value' or similar.
# Replace 'Value' with the actual name of your column if it's different.
plt.plot(df.index, df['Burglary/Breaking and Entering'], label='Observed') # Modified Line
plt.grid(True)
plt.legend() # Added a legend to distinguish the lines
plt.show()
```
