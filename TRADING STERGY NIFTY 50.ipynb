```python
# Install required packages
!pip install pandas numpy scikit-learn yfinance ta

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')

def load_and_prepare_data():
    # Upload the file through Colab's interface
    from google.colab import files
    uploaded = files.upload()

    # Get the filename of the uploaded file
    filename = list(uploaded.keys())[0]

    # Read the CSV file
    df = pd.read_csv(filename)

    # Convert date column to datetime
    df['Datetime'] = pd.to_datetime(df['date'])
    df.set_index('Datetime', inplace=True)

    # Handle any missing values in core columns
    core_columns = ['open', 'high', 'low', 'close', 'volume']
    df[core_columns] = df[core_columns].fillna(method='ffill')

    return df

def add_technical_features(df):
    try:
        # Price-based features
        df['Returns'] = df['close'].pct_change().fillna(0)
        df['Log_Returns'] = np.log(df['close']/df['close'].shift(1)).fillna(0)

        # Moving averages
        for w in [5, 10, 20]:
            df[f'MA_{w}'] = df['close'].rolling(window=w).mean().fillna(method='bfill')
            df[f'MA_Ratio_{w}'] = (df['close'] / df[f'MA_{w}']).fillna(1)

        # Volatility features
        df['Daily_Range'] = (df['high'] - df['low']).fillna(0)
        df['Range_MA_5'] = df['Daily_Range'].rolling(window=5).mean().fillna(method='bfill')

        # Volume features
        df['Volume_MA_5'] = df['volume'].rolling(window=5).mean().fillna(method='bfill')
        df['Volume_Ratio'] = (df['volume'] / df['Volume_MA_5']).fillna(1)

        # RSI with error handling
        delta = df['close'].diff()
        gain = delta.where(delta > 0, 0).fillna(0)
        loss = (-delta.where(delta < 0, 0)).fillna(0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = (avg_gain / avg_loss).fillna(0)
        df['RSI'] = 100 - (100 / (1 + rs))
        df['RSI'] = df['RSI'].fillna(50)  # Fill RSI NaN with neutral value

        # Bollinger Bands
        df['BB_Middle'] = df['close'].rolling(window=20).mean().fillna(method='bfill')
        bb_std = df['close'].rolling(window=20).std().fillna(method='bfill')
        df['BB_Upper'] = df['BB_Middle'] + 2*bb_std
        df['BB_Lower'] = df['BB_Middle'] - 2*bb_std
        df['BB_Position'] = ((df['close'] - df['BB_Lower']) / (df['BB_Upper'] - df['BB_Lower'])).fillna(0.5)

        # Momentum features
        df['Momentum'] = (df['close'] - df['close'].shift(5)).fillna(0)
        df['Rate_of_Change'] = df['close'].pct_change(periods=5).fillna(0)

        return df

    except Exception as e:
        print(f"Error in feature engineering: {str(e)}")
        raise

def create_target(df, threshold=0.001):
    df['Target'] = ((df['close'].shift(-1) - df['close']) / df['close'] > threshold).astype(int)
    df['Target'] = df['Target'].fillna(0)  # Fill last row
    return df

def prepare_features(df):
    feature_columns = [col for col in df.columns if col not in ['date', 'Target', 'open', 'high', 'low', 'close', 'volume']]

    print("Shape before feature preparation:", df.shape)
    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.fillna(method='ffill')
    df = df.fillna(method='bfill')  # Handle any remaining NaNs
    print("Shape after feature preparation:", df.shape)

    X = df[feature_columns]
    y = df['Target']
    return X, y

def train_model(X, y):
    try:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

        model = RandomForestClassifier(
            n_estimators=100,
            max_depth=10,
            min_samples_split=20,
            min_samples_leaf=10,
            random_state=42,
            n_jobs=-1  # Use all available cores
        )

        model.fit(X_train, y_train)

        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)

        train_accuracy = accuracy_score(y_train, train_pred)
        test_accuracy = accuracy_score(y_test, test_pred)

        return model, train_accuracy, test_accuracy, X_test, y_test

    except Exception as e:
        print(f"Error in model training: {str(e)}")
        raise

def main():
    try:
        print("Upload your NIFTY data CSV file:")
        df = load_and_prepare_data()

        print("Adding technical features...")
        df = add_technical_features(df)

        print("Creating target variable...")
        df = create_target(df)

        print("Preparing features...")
        X, y = prepare_features(df)

        print("Training model...")
        model, train_acc, test_acc, X_test, y_test = train_model(X, y)

        print(f"\nTraining Accuracy: {train_acc:.3f}")
        print(f"Test Accuracy: {test_acc:.3f}")

        feature_importance = pd.DataFrame({
            'feature': X.columns,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)

        print("\nTop 10 Most Important Features:")
        print(feature_importance.head(10))

        return model, feature_importance

    except Exception as e:
        print(f"Error in main execution: {str(e)}")
        raise

if __name__ == "__main__":
    model, feature_importance = main()
```

```python
import pandas as pd
import numpy as np

class NiftyTradingStrategy:
    def __init__(self, initial_capital=10000):
        self.capital = initial_capital
        self.positions = {}
        self.trades = []

    def calculate_features(self, df):
        # Most important features based on model results
        df['Range_MA_5'] = (df['high'] - df['low']).rolling(window=5).mean()
        df['MA_20'] = df['close'].rolling(window=20).mean()
        df['MA_5'] = df['close'].rolling(window=5).mean()
        df['MA_10'] = df['close'].rolling(window=10).mean()

        # Bollinger Bands
        df['BB_Middle'] = df['close'].rolling(window=20).mean()
        df['BB_std'] = df['close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_Middle'] + 2 * df['BB_std']
        df['BB_Lower'] = df['BB_Middle'] - 2 * df['BB_std']

        # RSI
        delta = df['close'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
        rs = gain / loss
        df['RSI'] = 100 - (100 / (1 + rs))

        return df.dropna()

    def generate_signals(self, df):
        """
        Generate trading signals based on the most important features
        1: Buy signal
        -1: Sell signal
        0: No signal
        """
        signals = pd.Series(0, index=df.index)

        # Entry conditions (composite strategy based on top features)
        buy_conditions = (
            (df['close'] > df['MA_5']) &  # Price above short-term MA
            (df['close'] > df['BB_Lower']) &  # Price above lower BB
            (df['Range_MA_5'] < df['Range_MA_5'].rolling(window=20).mean()) &  # Volatility decreasing
            (df['RSI'] > 40) & (df['RSI'] < 70)  # RSI in moderate range
        )

        # Exit conditions
        sell_conditions = (
            (df['close'] < df['MA_5']) |  # Price below short-term MA
            (df['close'] > df['BB_Upper']) |  # Price above upper BB
            (df['RSI'] > 70) |  # Overbought
            (df['RSI'] < 40)  # Oversold
        )

        signals[buy_conditions] = 1
        signals[sell_conditions] = -1

        return signals

    def backtest(self, df, position_size=0.2):
        """
        Backtest the strategy
        position_size: fraction of capital to invest per trade
        """
        df = self.calculate_features(df)
        df['Signal'] = self.generate_signals(df)

        portfolio_value = []
        current_position = 0
        entry_price = 0

        for i in range(len(df)):
            current_price = df['close'].iloc[i]
            signal = df['Signal'].iloc[i]

            # Handle position entry
            if signal == 1 and current_position == 0:
                shares = (self.capital * position_size) / current_price
                entry_price = current_price
                current_position = shares
                self.trades.append({
                    'date': df.index[i],
                    'type': 'BUY',
                    'price': current_price,
                    'shares': shares
                })

            # Handle position exit
            elif signal == -1 and current_position > 0:
                profit = (current_price - entry_price) * current_position
                self.capital += profit
                self.trades.append({
                    'date': df.index[i],
                    'type': 'SELL',
                    'price': current_price,
                    'shares': current_position,
                    'profit': profit
                })
                current_position = 0

            # Track portfolio value
            portfolio_value.append(self.capital + (current_position * current_price))

        df['Portfolio_Value'] = portfolio_value
        return df
```

```python
!pip install seaborn
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

class StrategyAnalyzer:
    def __init__(self, results_df, trades):
        self.results_df = results_df
        self.trades = trades
        self.start_date = results_df.index[0]
        self.end_date = results_df.index[-1]
        self.start_value = results_df['Portfolio_Value'].iloc[0]
        self.end_value = results_df['Portfolio_Value'].iloc[-1]

    def calculate_cagr(self):
        """Calculate Compound Annual Growth Rate"""
        num_years = (self.end_date - self.start_date).days / 365.25
        cagr = (self.end_value / self.start_value) ** (1 / num_years) - 1
        return cagr

    def calculate_xirr(self):
        """Calculate XIRR using trade cashflows"""
        try:
            cashflows = []
            # Add initial investment
            cashflows.append((self.start_date, -self.start_value))

            # Add all trades
            for trade in self.trades:
                amount = trade['price'] * trade['shares']
                if trade['type'] == 'BUY':
                    amount = -amount
                cashflows.append((trade['date'], amount))

            # Add final portfolio value
            cashflows.append((self.end_date, self.end_value))

            # Convert dates to numbers for numpy
            dates = [(date - self.start_date).days for date, _ in cashflows]
            amounts = [amount for _, amount in cashflows]

            return np.irr(amounts)
        except:
            return None

    def calculate_monthly_metrics(self):
        """Calculate monthly profit/loss and other metrics"""
        self.results_df['Daily_Returns'] = self.results_df['Portfolio_Value'].pct_change()
        monthly_returns = self.results_df['Daily_Returns'].resample('M').apply(
            lambda x: (1 + x).prod() - 1
        )

        return {
            'Monthly_Returns': monthly_returns,
            'Avg_Monthly_Return': monthly_returns.mean(),
            'Monthly_Volatility': monthly_returns.std(),
            'Sharpe_Ratio': (monthly_returns.mean() * 12) / (monthly_returns.std() * np.sqrt(12))
        }

    def plot_performance(self):
        """Create performance visualization"""
        sns.set_style("whitegrid")
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

        # Portfolio Value
        sns.lineplot(data=self.results_df, x=self.results_df.index, y='Portfolio_Value', ax=ax1)
        ax1.set_title('Portfolio Value Over Time')
        ax1.set_ylabel('Portfolio Value')

        # Monthly Returns
        monthly_metrics = self.calculate_monthly_metrics()
        monthly_returns = monthly_metrics['Monthly_Returns']
        sns.barplot(x=monthly_returns.index.astype(str), y=monthly_returns.values * 100, ax=ax2)
        ax2.set_title('Monthly Returns (%)')
        ax2.set_ylabel('Return (%)')
        ax2.tick_params(axis='x', rotation=45)

        plt.tight_layout()
        return fig

    def generate_report(self):
        """Generate comprehensive performance report"""
        cagr = self.calculate_cagr()
        xirr = self.calculate_xirr()
        monthly_metrics = self.calculate_monthly_metrics()

        report = {
            'CAGR (%)': cagr * 100,
            'XIRR (%)': xirr * 100 if xirr else None,
            'Total Return (%)': ((self.end_value / self.start_value) - 1) * 100,
            'Annualized Volatility (%)': monthly_metrics['Monthly_Volatility'] * np.sqrt(12) * 100,
            'Sharpe Ratio': monthly_metrics['Sharpe_Ratio'],
            'Number of Trades': len(self.trades),
            'Average Monthly Return (%)': monthly_metrics['Avg_Monthly_Return'] * 100,
            'Max Drawdown (%)': self.calculate_max_drawdown() * 100
        }

        return pd.Series(report)

    def calculate_max_drawdown(self):
        """Calculate maximum drawdown"""
        rolling_max = self.results_df['Portfolio_Value'].expanding().max()
        drawdowns = self.results_df['Portfolio_Value'] / rolling_max - 1
        return abs(drawdowns.min())

def main():
    analyzer = StrategyAnalyzer(results_df, strategy.trades)
    report = analyzer.generate_report()
    print("\nStrategy Performance Report:")
    print(report.to_string())

    fig = analyzer.plot_performance()
    plt.show()

if __name__ == "__main__":
    main()
```
